# Economics Simulation

This is a simulation of an economic marketplace in which there is a *population* of actors, each of which has a level of wealth (a single number) that changes over time.  On each time step two agents (chosen by an *interaction rule*) interact with each other and exchange wealth (according to a *transaction rule*).  The idea is to understand the evolution of the population's wealth over time.  My hazy memory is that this idea came from a class by Prof. [Sven Anderson](http://turing.bard.edu/~sven/) at Bard (any errors or misconceptions here are due to my (Peter Norvig) misunderstanding of his idea). Why this is interesting: (1) an example of using simulation to model the world. (2) Many students will have preconceptions about how economies work that will be challenged by the results shown here.

## Population Distributions

First things first: what should our initial population look like?  We will provide several distribution functions (constant, uniform, Gaussian, etc.) and a `sample` function, which samples *N* elements form a distribution and then normalizes them to have a given mean.  By default we will have *N*=5000 actors and an initial mean wealth of 100 simoleons.

```python
import random
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

N  = 5000 # Default size of population
mu = 100. # Default mean of population's wealth
total_wealth = N * mu

def sample(distribution, N=N, mu=mu):
    "Sample from the distribution N times, then normalize results to have mean mu."
    return normalize([distribution() for _ in xrange(N)], total_wealth)

def constant(mu=mu):          return mu
def uniform(mu=mu, width=mu): return random.uniform(mu-width/2, mu+width/2)
def gauss(mu=mu, sigma=mu/3): return random.gauss(mu, sigma) 
def beta(alpha=2, beta=3):    return random.betavariate(alpha, beta)
def pareto(alpha=4):          return random.paretovariate(alpha)
    
def normalize(numbers, total):
    "Scale the numbers so that they add up to total."
    factor = total / float(np.sum(numbers))
    return factor * np.array(numbers)

#def normalize(numbers, total):
#    "Scale the numbers so that they add up to total."
#    factor = total / float(sum(numbers))
#    return np.array([x * factor for x in numbers])
```

## Transactions

In a transaction, two actors come together; they have existing wealth levels X and Y.  For now we will only consider transactions that conserve wealth, so our transaction rules will decide how to split up the pot of X+Y total wealth.

```python
def random_split(X, Y):
    "Take all the money in the pot and divide it randomly between X and Y."
    pot = X + Y
    m = random.uniform(0, pot)
    return m, pot - m
    
def winner_take_most(X, Y, most=3/4.): 
    "Give most of the money in the pot to one of the parties."
    pot = X + Y
    m = random.choice((most * pot, (1 - most) * pot))
    return m, pot - m

def winner_take_all(X, Y): 
    "Give all the money in the pot to one of the actors."
    return winner_take_most(X, Y, 1.0)

def redistribute(X, Y): 
    "Give 55% of the pot to the winner; 45% to the loser."
    return winner_take_most(X, Y, 0.55)

def split_half_min(X, Y):
    """The poorer actor only wants to risk half his wealth; 
    the other actor matches this; then we randomly split the pot."""
    pot = min(X, Y)
    m = random.uniform(0, pot)
    return X - pot/2. + m, Y + pot/2. - m

def split_half_min_improved(X, Y):
    """The poorer actor only wants to risk half his wealth; 
    the other actor matches this; then we randomly split the pot."""
    pot = 0.5 * min(X, Y)
    m = random.uniform(-pot, pot)
    return X + m, Y - m

def split_frac_min(X, Y, frac=0.1):
    """The poorer actor only wants to risk half his wealth; 
    the other actor matches this; then we randomly split the pot."""
    pot = frac * min(X, Y)
    m = random.uniform(-pot, pot)
    return X + m, Y - m

def sfm_gauss(X, Y, frac=0.1, luck=0.5):
    """The poorer actor only wants to frac of his wealth; 
    the other actor matches this; then we randomly split the pot
    according to a Gaussian distribution."""
    max_wealth_poor = min(X, Y)
    transaction_value = frac*max_wealth_poor
    while True:
        m = random.gauss(0, luck*transaction_value)
        if abs(m) <= max_wealth_poor: break
    return X + m, Y - m

def sfm_gauss_mod(X, Y, frac=0.1, luck=0.5):
    """The poorer actor risks frac of his wealth on average; the
    distribution of wealth committed to pot is modeled by a distribution;
    the wealthier actor matches this commitment. The pot is split with
    a Gaussian describing the probability that the poorer actor gains or
    loses wealth and vice versa for the wealthier actor."""
    max_wealth_poor = min(X, Y)
    transaction_value = np.random.exponential(frac*max_wealth_poor)
    while True:
        m = random.gauss(0, luck*transaction_value)
        if abs(m) <= max_wealth_poor: break
    return X + m, Y - m
```

## Interactions

How do you decide which parties interact with each other?  The rule `anyone` samples two members of the population uniformly and independently, but there are other possible rules, like `nearby(pop, k)`, which choses one member uniformly and then chooses a second within `k` index elements away, to simulate interactions within a local neighborhood.

```python
def anyone(pop): return random.sample(range(len(pop)), 2)

def nonlocal(pop): return np.random.randint(0, high=len(pop), size=2)

def nearby(pop, k=5): 
    i = random.randrange(len(pop))
    j = i + random.choice((1, -1)) * random.randint(1, k)
    return i, (j % len(pop))
               
def nearby1(pop): return nearby(pop, 1)
```

## Simulation

Now let's describe the code to run the simulation and summarize/plot the results.  The function `simulate` does the work; it runs the interaction function to find two actors, then calls the transaction function to figure out how to split their wealth, and repeats this T times. The only other thing it does is record results.  Every so-many steps, it records some summary statistics of the population (by default, this will be every 25 steps).

What information do we record to summarize the population?  Out of the N=5000 (by default) actors, we will record the wealth of exactly nine of them: the ones, in sorted-by-wealth order that occupy the 1% spot (that is, if N=5000, this would be the 50th wealthiest actor), then the 10%, 25% 1/3, and median; and then likewise from the bottom the 1%, 10%, 25% and 1/3.

(Note that we record the *median*, which changes over time; the *mean* is defined to be 100 when we start, and since all transactions conserve wealth, the mean will always be 100.)

What do we do with these results, once we have recorded them?  First we print them in a table for the first time step, the last, and the middle.  Then we plot them as nine lines in a plot where the y-axis is wealth and the x-axis is time (note that when the x-axis goes from 0 to 1000, and we have `record_every=25`, that means we have actually done 25,000 transactions, not 1000).

```python
def simulate(population, transaction_fn, interaction_fn, T, percentiles, record_every,      \
             wealth_tax, tax_freq, verbose):
    "Run simulation for T steps; collect percentiles every 'record_every' time steps."
    results = []
    detailed = []
    public_wealth = wealth_tax*total_wealth # tax
    for t in xrange(T):
        i, j = interaction_fn(population)
        population[i], population[j] = transaction_fn(population[i], population[j]) 
        if t % record_every == 0:
            if verbose: print 'Recording time step %i...' % t
            results.append(record_percentiles(population, percentiles))
            detailed.append(population)
            if wealth_tax and t % int(T*tax_freq) == 0:
                if verbose: print '  --> redistributing wealth'
                population = (1.0 - wealth_tax) * population
                population += public_wealth / N # redistribute wealth
    return results, detailed

def report(distribution=constant, transaction_fn=random_split, interaction_fn=nonlocal,    \
           N=N, mu=mu, T=5*N, percentiles=(1, 10, 25, 33.3, 50, -33.3, -25, -10, -1),      \
           record_every=25, wealth_tax=0.0, tax_freq=0.1, verbose=False):
    "Print and plot the results of the simulation running T steps." 
    # Run simulation
    population = sample(distribution, N, mu)
    results, detailed = simulate(population, transaction_fn, interaction_fn, T, percentiles,    \
                                 record_every, wealth_tax, tax_freq, verbose)
    # Print summary
    print('Simulation: {} * {}(mu={}) for T={} steps with {} interactions and {} transactions:\n'.format(
          N, name(distribution), mu, T, name(interaction_fn), name(transaction_fn)))
    fmt = '{:6}' + '{:10.2f} ' * len(percentiles)
    print(('{:6}' + '{:>10} ' * len(percentiles)).format('', *map(percentile_name, percentiles)))
    for (label, nums) in [('start', results[0]), ('mid', results[len(results)//2]), ('final', results[-1])]:
        print fmt.format(label, *nums)
    # Plot results
    for line in zip(*results):
        plt.plot(line)
    plt.show()
    
    return results, detailed

def record_percentiles(population, percentiles):
    "Pick out the percentiles from population."
    population = sorted(population, reverse=True)
    N = len(population)
    return [population[int(p*N/100.)] for p in percentiles]

def percentile_name(p):
    return ('median' if p == 50 else 
            '{} {}%'.format(('top' if p > 0 else 'bot'), abs(p)))
    
def name(obj):
    return getattr(obj, '__name__', str(obj))
```

## Effect of transaction behavior and wealth taxation

The poorer of the two actors in a transaction commits a fraction of their wealth to the pot. The wealthier actor commits the same amount. For any given transaction, neither actor will gain or lose wealth on average ("you get what you pay for"); however there is a finite probability that one actor "gets more than they paid for" (a bonus of +m simoleons to their wealth), while the other loses the same amount (a deduction of m simoleons from their wealth). Transation bonuses are represented as a Gaussian distribution about transaction parity (zero mean) with a width determining the likelihood larger/smaller bonuses occur. If the bonus for the wealthier actor is greater than the total wealth of the poorer actor, then the wealtheir actor acquires all of the poorer actor's wealth.

### Split quarter min

The poorer of the two actors in a transaction commits a **quarter** of their wealth to the pot.

```python
def split_quarter_min(X, Y): return split_frac_min(X, Y, frac=0.25)
results1, detailed1 = report(constant, split_quarter_min, T=50*N)
results2, detailed2 = report(constant, split_quarter_min, T=50*N, wealth_tax=0.1)
```

### Split tenth min with intelligent spending

The poorer of the two actors in a transaction commits, on average, a **tenth** of their wealth to the pot. However, the value of any given transaction is exponentially (I will use Poisson later since it seems more appropriate) distributed about the mean spending of the poorer actor; the maximum amount that can be committed by each actor to the pot is limited to the maximum wealth of the poorer actor (no debt allowed). This models the idea that smaller transactions should be more frequent than larger transactions.
